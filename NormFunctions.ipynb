## Generating vector data

### Subtask:
Create a simple vector to demonstrate the calculation of different norms.


**Reasoning**:
The instructions are to create a simple vector using NumPy. This requires importing the `numpy` library and then creating a NumPy array.



```python
import numpy as np

# Create a simple vector
vector = np.array([3, 4])

# Display the vector
print("The simple vector is:")
print(vector)
```

## Calculating different norms

### Subtask:
Write code to calculate L1, L2, and L-infinity norms of the generated vector.


**Reasoning**:
Calculate and print the L1, L2, and L-infinity norms of the generated vector using numpy.linalg.norm.



```python
# Calculate the L1 norm
l1_norm = np.linalg.norm(vector, ord=1)

# Calculate the L2 norm
l2_norm = np.linalg.norm(vector, ord=2) # ord=2 is the default

# Calculate the L-infinity norm
l_inf_norm = np.linalg.norm(vector, ord=np.inf)

# Print the calculated norms
print(f"L1 Norm: {l1_norm}")
print(f"L2 Norm: {l2_norm}")
print(f"L-infinity Norm: {l_inf_norm}")
```

## Visualizing norms

### Subtask:
Create visualizations to intuitively understand the difference between these norms. This could involve plotting points on a 2D plane and showing the different "shapes" of the unit balls for each norm.


**Reasoning**:
The subtask requires creating visualizations of the unit balls for L1, L2, and L-infinity norms. This involves generating points that satisfy the unit norm condition for each type and then plotting these points. The provided steps outline the process of generating points and plotting for each norm. Since plotting all three unit balls together in a single figure provides a clearer comparison, I will group the generation and plotting for all three norms into one code block. I will start by importing the required library, `matplotlib.pyplot`, and then generate the points for each norm's unit ball and plot them.



```python
import matplotlib.pyplot as plt
import numpy as np

# Generate points for L1 unit ball (|x| + |y| = 1)
theta = np.linspace(0, 2*np.pi, 1000)
x_l1 = np.cos(theta)
y_l1 = np.sin(theta)
# Transform points to fit the L1 unit ball shape
x_l1_unit = np.sign(x_l1) * np.minimum(np.abs(x_l1) + np.abs(y_l1), 1) * np.abs(x_l1) / (np.abs(x_l1) + np.abs(y_l1) + 1e-9)
y_l1_unit = np.sign(y_l1) * np.minimum(np.abs(x_l1) + np.abs(y_l1), 1) * np.abs(y_l1) / (np.abs(x_l1) + np.abs(y_l1) + 1e-9)
# A simpler way to generate L1 unit ball points:
x_l1_simple = np.concatenate([np.linspace(1, 0, 100), np.linspace(0, -1, 100), np.linspace(-1, 0, 100), np.linspace(0, 1, 100)])
y_l1_simple = np.concatenate([np.linspace(0, 1, 100), np.linspace(1, 0, 100), np.linspace(0, -1, 100), np.linspace(-1, 0, 100)])


# Generate points for L2 unit ball (x^2 + y^2 = 1)
theta_l2 = np.linspace(0, 2*np.pi, 1000)
x_l2 = np.cos(theta_l2)
y_l2 = np.sin(theta_l2)

# Generate points for L-infinity unit ball (max(|x|, |y|) = 1)
x_linf = np.concatenate([np.linspace(-1, 1, 100), np.ones(100), np.linspace(1, -1, 100), -np.ones(100)])
y_linf = np.concatenate([-np.ones(100), np.linspace(-1, 1, 100), np.ones(100), np.linspace(1, -1, 100)])


# Create the plot
plt.figure(figsize=(8, 8))

plt.plot(x_l1_simple, y_l1_simple, label='L1 Norm (Manhattan Distance)')
plt.plot(x_l2, y_l2, label='L2 Norm (Euclidean Distance)')
plt.plot(x_linf, y_linf, label='L-infinity Norm (Chebyshev Distance)')

plt.axhline(0, color='grey', lw=0.5)
plt.axvline(0, color='grey', lw=0.5)
plt.grid(True, linestyle='--', alpha=0.6)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Unit Balls for Different Norms')
plt.axis('equal')
plt.legend()
plt.show()
```
